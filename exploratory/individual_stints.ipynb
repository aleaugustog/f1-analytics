# Individual Session & Driver Stints

```python
import pandas as pd

meetings_df = pd.read_json('https://api.openf1.org/v1/meetings?year=2025')
meetings_df
```

## Sessions

```python
sessions_df = pd.read_json('https://api.openf1.org/v1/sessions?meeting_key=1277')
sessions_df
```

## Set Session & Driver

```python
session_key = 9941
driver_number = 31
```

## Stints

```python
stints_df = pd.read_json(f'https://api.openf1.org/v1/stints?session_key={session_key}&driver_number={driver_number}')

compound_to_color = {
    "SOFT": "#FF0000",   # red
    "MEDIUM": "#FFFF00", # yellow
    "HARD": "#FFFFFF",   # white
}

# Map compound colors to the stints DataFrame
stints_df["compound_color"] = stints_df["compound"].map(compound_to_color)

stints_df
```

## Laps

```python
laps_df = pd.read_json(f'https://api.openf1.org/v1/laps?session_key={session_key}&driver_number={driver_number}')
laps_df = laps_df[['session_key', 'driver_number', 'lap_number', 'is_pit_out_lap', 'lap_duration']]
laps_df
```

### Matching each lap to the corresponding stint

```python
# Matching each lap to the corresponding stint on a new DF
laps_stints_df = laps_df.merge(stints_df, on=['session_key', 'driver_number'])
laps_stints_df = laps_stints_df[(laps_stints_df['lap_number'] >= laps_stints_df['lap_start']) & (laps_stints_df['lap_number'] <= laps_stints_df['lap_end']) & (laps_stints_df['is_pit_out_lap'] == False)]
laps_stints_df
```

```python
notna_laps_df = laps_stints_df[laps_stints_df['lap_duration'].notna()]
notna_laps_df['lap_duration'].describe()
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is called df and has columns:
# 'lap_number', 'lap_duration', 'compound_color'

plt.figure(figsize=(12, 6), facecolor='black')

ax = plt.gca()
ax.set_facecolor('black')

plt.bar(
    x=notna_laps_df["lap_number"],
    height=notna_laps_df["lap_duration"],
    color=notna_laps_df["compound_color"]
)

ax.tick_params(colors='white')  # Tick labels
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
plt.xlabel("Lap Number", color='white')
plt.ylabel("Lap Duration (s)", color='white')
plt.title("Lap Duration by Lap", color='white')
plt.grid(True, axis='y', linestyle='--', alpha=0.5, color='white')

plt.tight_layout()
plt.show()
```
